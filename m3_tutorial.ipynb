```python
class Solution:
    def sort012(self, arr):
        zero = []
        one = []
        two = []
        
        for a in arr:
            if a == 0:
                zero.append(a)
            elif a == 1:
                one.append(a)
            else:
                two.append(a)
        
        final = zero + one + two
        
        for i in range(len(final)):
            arr[i] = final[i]
            
        return arr
        
if __name__ == '__main__':
    sol = Solution()
    
    arr = [0, 1, 2, 0, 1, 2]
    
    sol.sort012(arr)
```

```python
class Solution:
    def binarysearch(self, arr, k):
        low = 0
        high = len(arr) - 1
        
        while low <= high:
            mid = (low + high)//2
            
            if arr[mid] == k:
                while arr[mid-1] == k:
                    mid -= 1
                return mid
            elif arr[mid] < k:
                low = mid + 1
            else:
                high = mid - 1
        
        return -1
        
if __name__ == '__main__':
    sol = Solution()
    arr = [1, 2, 3, 4, 5]
    k = 4
    
    sol.binarysearch(arr, k)
```
